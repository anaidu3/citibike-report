```python
# Combining all csv files, since no unique join column id
df = df_feb_2021.append(df_feb_2021, sort=False)
df = df.append(df_mar_2021, sort=False)
df = df.append(df_apr_2021, sort=False)
df = df.append(df_may_2021, sort=False)
df = df.append(df_jun_2021, sort=False)
df = df.append(df_jul_2021, sort=False)
df = df.append(df_aug_2021, sort=False)
df = df.append(df_sept_2021, sort=False)
df = df.append(df_oct_2021, sort=False)
df = df.append(df_nov_2021, sort=False)
df = df.append(df_dec_2021, sort=False)
df = df.append(df_jan_2022, sort=False)
df = df.append(df_feb_2022, sort=False)
df = df.append(df_mar_2022, sort=False)
df = df.append(df_apr_2022, sort=False)
df = df.append(df_may_2022, sort=False)
df = df.append(df_jun_2022, sort=False)
```

```python
df['ended_at'] = pd.to_datetime(df.ended_at)
```

```python
df['started_at'] = pd.to_datetime(df.started_at)
```

```python
df['trip_duration_min'] = (df['ended_at'] - df['started_at']).dt.total_seconds() / 60
```
